```python
import numpy as np
```

```python
# 30fpsと仮定して
fps = 30.0
freq = 1

pi = np.arccos(-1)
f = open('sin'+str(freq)+'.dat', 'w') # 書き込みモードで開く

for i in range(0,1000):
    data = np.sin(2*pi*freq*i/fps)
    string = str(i)+" "+str(data)+"\n"
    print(string)
    f.write(string)

f.close()
```
